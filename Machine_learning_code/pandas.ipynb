```python
train_data.dropna(subset=['Age'])
train_data.dropna(axis=1)
# na가 하나라도 있으면 칼럼 제거
```

```python
m1 = train_data[train_data['Survived']==1]['Age'].mean()
m2 = train_data[train_data['Survived']==0]['Age'].mean()
train_data.loc[train_data['Survived']==1,'Age']= train_data[train_data['Survived']==1]['Age'].fillna(m1)
train_data.loc[train_data['Survived']==0,'Age'] = train_data[train_data['Survived']==0]['Age'].fillna(m2)
```

```python
import math
def age_cate(age):
    return math.floor(age/10)*10
train_data['Age_cate'] = train_data['Age'].apply(age_cate)
train_data
```

```python
#pd.get_dummies(train_data, columns=['Pclass'], drop_first =True)
#첫번째 변수는 필요가 없다.
class_group = train_data.groupby('Pclass')
class_group.groups
```

```python
train_data.groupby(['Pclass','Sex']).mean()['Survived']
```

```python
import numpy as np
train_data.set_index('Age_cate').groupby(level=0).mean()
train_data.set_index(['Pclass','Sex']).groupby(level=[0,1]).mean()['Age']
train_data.set_index(['Pclass','Sex']).groupby(level=[0,1]).aggregate([np.mean,np.sum,np.max])
```

```python
train_data.groupby('Pclass').transform(np.mean)['Age']
train_data.groupby(['Pclass','Sex']).transform(np.mean)['Age']
# 기존의 인덱스와 통계함수를 유지, 새로운 결과를 합칠수 있다.
```
