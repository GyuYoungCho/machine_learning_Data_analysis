```python
from collections import Counter

train_index = (c4["applied"]!="unknown")
test_index = (c4["applied"]=="unknown")

train_index = np.array(train_index)
test_index = np.array(test_index)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import OneHotEncoder
nb = MultinomialNB()

from sklearn.compose import ColumnTransformer

cat_attribs = ["keyword_x","keyword_y","companySize"]

full_pipeline = ColumnTransformer([
        ("encoder", OneHotEncoder(categories='auto'), cat_attribs),
    ])
prepared = full_pipeline.fit_transform(c4)
prepared_train = prepared[train_index,:]
prepared_test = prepared[test_index,:]
c4_y = c4["applied"]
y_train = c4_y.iloc[train_index]
y_test = c4_y.iloc[test_index]
y_train = y_train.astype('int')
```
